```python
image_path = "/content/city.jpg"
image = Image.open(image_path)

display(image)

image_array = np.array(image)

h, w = image_array.shape[:2]

modified_image_array = np.array(image_array)

for x in range(1, h - 1):
    for y in range(1, w - 1):
        modified_pixel_value = (
            (image_array[x + 1][y] +
            image_array[x - 1][y] +
            image_array[x][y + 1] +
            image_array[x][y - 1])
        ) //4
        modified_image_array[x][y] = modified_pixel_value

modified_image = Image.fromarray(np.uint8(modified_image_array))

display(modified_image)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image from a file (replace "image_path.jpg" with your actual image path)
image_path = "/content/padmabridge.jpg"
image = Image.open(image_path)

# Convert the image to grayscale
grayscale_image = image.convert("L")

# Invert the image
inverted_image = Image.eval(grayscale_image, lambda x: 255 - x)

# Display the inverted image using matplotlib
plt.imshow(inverted_image, cmap='gray')
plt.axis('off')  # Turn off axis ticks and labels
plt.show()
```

```python
from IPython.testing import test
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the grayscale image
image_path = "/content/sample_data/apple.jpg"  # Replace with your image path
input_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if input_image is None:
    print("Error: Unable to load image")
else:
    # Apply Sobel edge detection
    sobelx = cv2.Sobel(input_image, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(input_image, cv2.CV_64F, 0, 1, ksize=3)

    # Calculate the gradient magnitude
    gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)

    # Normalize and convert to 8-bit for visualization
    gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

    # Display the original and edge-detected images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(input_image, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Sobel Edge Detection")
    plt.axis('off')

    plt.tight_layout()


    plt.imshow(gradient_magnitude, cmap='gray')


    plt.show()

```
